```python
import glob
import nltk
import string

def bigram_collo(tokens,window_size=2,num_results=50):
    """
    Creates a collocation object based on tokens and a couple of other parameters
    """
    # define statistical tests for bigrams
    bigram_measures = nltk.collocations.BigramAssocMeasures()
    # go and find bigrams
    finder = nltk.collocations.BigramCollocationFinder.from_words(tokens, window_size)
    ignored_words = nltk.corpus.stopwords.words('english')
    finder.apply_word_filter(lambda w: w.lower() in ignored_words or not w.isalnum())
    # measure which bigrams are important and print the top 50
    result = sorted(finder.nbest(bigram_measures.raw_freq, num_results))
    return result
```

```python
for f in glob.glob("TEXT/*.txt"):
    print("\n",f[5:-4],"\n")
    bc = bigram_collo(nltk.tokenize.word_tokenize(open(f).read()))
    for b in bc:
        print(b)
```

```python
lemmatizer  = nltk.WordNetLemmatizer()
stopwords   = set(nltk.corpus.stopwords.words('english'))
punctuation = string.punctuation

def normalize(text):
    """
    Returns a token
    """
    for token in nltk.word_tokenize(text):
        token = token.lower()
        token = lemmatizer.lemmatize(token)
        if token not in stopwords and token not in punctuation:
            return token
        
def normalize_wordlist(wordlist):
    """
    Completed Normalized Wordlist as tokens? Takes Sentenances, eg. sent_tokenizer.tokenize()
    """
    normalized_wordlist = [normalize(w) for w in wordlist]
    # discard punctuation
    return [tok for tok in normalized_wordlist if tok is not None]
```

```python
def bigram_collo_normalized(tokens,window_size=2,num_results=50):
    """
    Creates a collocation object based on tokens and a couple of other parameters
    """
    # define statistical tests for bigrams
    bigram_measures = nltk.collocations.BigramAssocMeasures()
    # go and find bigrams
    finder = nltk.collocations.BigramCollocationFinder.from_words(tokens, window_size)
    ignored_words = nltk.corpus.stopwords.words('english')
    finder.apply_word_filter(lambda w: w.lower() in ignored_words or not w.isalnum())
    # measure which bigrams are important and print the top 50
    result = sorted(finder.nbest(bigram_measures.raw_freq, num_results))
    return result
```

```python
sent_tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

for f in glob.glob("TEXT/*.txt"):
    print("\n",f[5:-4],"\n")
    raw = open(f).read()
    sents = sent_tokenizer.tokenize(raw)
    nwt = normalize_wordlist(sents)
    bc = bigram_collo_normalized(nwt,10)
    for b in bc:
        print(b)
```
