# Text Analysis for CAPAL paper

```python
%matplotlib inline
import pandas
google_data_url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSTFfUG61bAF56K_G01JTdoBqZil1l2IyNO0tohWQPA7TPzJetfvo0EZy4TYihMFfndvJCKLBxMeXpL/pub?gid=0&single=true&output=csv"
```

```python
ca_data = pandas.read_csv(google_data_url)
#print(ca_data.loc["Brock University"])
```

```python
pres_diff =(ca_data['prescribed_doctrine'].sum()/ca_data['prescribed_doctrine'].count())
ment_af =(ca_data['mentions_AF'].sum()/ca_data['mentions_AF'].count())
lib_lang =(ca_data['lib_specific_lang'].sum()/ca_data['lib_specific_lang'].count())
n_pres_diff = ca_data['prescribed_doctrine'].count()
n_ment_af = ca_data['mentions_AF'].count()
n_lib_lang =ca_data['lib_specific_lang'].count()



print('Doctrine {:.2%}'.format(pres_diff),"n = ",n_pres_diff)

print('Mentions AF {:.2%}'.format(ment_af),"n = ",n_ment_af)

print('Library Specific Language {:.2%}'.format(lib_lang),"n = ",n_lib_lang)
```

```python
mention = ['Yes', 'No']
series = pandas.Series([pres_diff,(1- pres_diff)], 
                   index=mention, 
                   name='Agreements that \'Without prescribed deference\'')
series.plot.pie(figsize=(6, 6))
```

```python
mention = ['Yes', 'No']
series = pandas.Series([ment_af,(1- ment_af)], 
                   index=mention, 
                   name='Agreements that Mention Academic Freedom')
series.plot.pie(figsize=(6, 6))
```

```python
mention = ['Specific Language', 'No Specific Language']
series = pandas.Series([lib_lang,(1- lib_lang)], 
                   index=mention, 
                   name='Librarian Specific Language on Academic Freedom')
series.plot.pie(figsize=(6, 6))
```
